## Stemming

The next couple of steps are meant to reduce words to their root word. Stemming achieves this by removing and suffixes which can reduce a word to a non-sensical root. Instinctually this may seem wrong but to the algorithm it is no different than a more natural language reduction. For example:

```python
stemmer = nltk.stem.SnowballStemmer("english")
print("The stemmed form of languages is: {}".format(stemmer.stem("languages")))
print("The stemmed form of language is: {}".format(stemmer.stem("language")))
```

There are some limits to Stemming though, reduction to a root without context can have some unintended consequences:

```python
print("The stemmed form of university is: {}".format(stemmer.stem("university")))
print("The stemmed form of universal is: {}".format(stemmer.stem("universal")))
```

## Lemmazation

Lemmazation tries to improve on the problem by having a dictionary of real words to verify the return. Instead of just chopping off the useless pieces it can deal with the more complex syntax that are inherient to english. For example, leaves and leaf are the same thing but with stemming it would not reduce to the same root. Lemmazation tries to help with that poblem by comparing it to a dictionary of known words. First we are going to take the Part of Speech work we already did and convert it to something that WordNet can utilize

```python
# Converts the nltk.pos_tag to POS tags used by WordNet
def get_wordnet_pos(x,treebank_tag):

    if treebank_tag.startswith('J'):
        return (x,wordnet.ADJ)
    elif treebank_tag.startswith('V'):
        return (x,wordnet.VERB)
    elif treebank_tag.startswith('N'):
        return (x,wordnet.NOUN)
    elif treebank_tag.startswith('R'):
        return (x,wordnet.ADV)
    elif treebank_tag.startswith('S'):
        return (x,wordnet.ADJ_SAT)
    else:
        return (x,wordnet.NOUN)
```

```python
pos = train['tokenPOS']
train['tokenPOS2'] = [[get_wordnet_pos(x[0],x[1]) for x in sentence] for sentence in pos]
```
